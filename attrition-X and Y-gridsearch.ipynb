```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy='most_frequent')
```

```python
mc=['Attrition', 'Age', 'BusinessTravel','Department', 'DistanceFromHome','EducationField','Gender','JobRole','MaritalStatus','MonthlyIncome', 'MonthlyRate','OverTime','PerformanceRating','StockOptionLevel']
for key in mc:   
    data[key]=si.fit_transform(data[[key]])
```

```python
data.isnull().sum()
```

```python
obj_list=['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(sparse= False)
```

```python
interim_op=ohe.fit_transform(data[obj_list])
```

```python
data_obj=pd.DataFrame(interim_op,columns=ohe.get_feature_names())
```

```python
data_obj
```

```python
data_obj.info()
```

```python
data_int=data.select_dtypes(include='number')
```

```python
data_int.info()
```

```python
data_final=pd.concat([data_obj,data_int],axis=1)
```

```python
data.shape
```

```python
data_final.shape
```

```python
data_final.info()
```

```python
data_final.isnull().sum()
```

```python
X=data_final
```

```python
X
```

```python
Y=data.loc[:,['Attrition']]
```

```python
Y
```

```python
Y.isnull().sum()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
param_grid={"max_depth":[5,10,15],"min_samples_split":[30,60,90],"criterion":["gini","entropy"]}
```

```python
Gd=GridSearchCV(model,param_grid,cv=10,n_jobs=-1,verbose=2,return_train_score= True)
```

```python
Gd.fit(X,Y)
```

```python
Gd.best_estimator_
```

```python
Gd.best_score_
```

```python
from sklearn.linear_model import LogisticRegression
```
